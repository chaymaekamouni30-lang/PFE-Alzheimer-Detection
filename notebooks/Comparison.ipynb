```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
cnn = pd.read_csv("/content/drive/MyDrive/pfe_fati_chaimae/CNN_Results.csv")

mobilenet = pd.read_csv("/content/drive/MyDrive/pfe_fati_chaimae/MobileNetV2_Results.csv")

```

Créer le tableau de comparaison

```python
comparison = pd.DataFrame({

    "Model":[
        "CNN",
        "MobileNetV2"
    ],

    "Accuracy":[
        cnn["Accuracy"][0],
        mobilenet["Accuracy"][0]
    ],

    "Precision":[
        cnn["Precision"][0],
        mobilenet["Precision"][0]
    ],

    "Recall":[
        cnn["Recall"][0],
        mobilenet["Recall"][0]
    ],

    "F1 Score":[
        cnn["F1 Score"][0],
        mobilenet["F1 Score"][0]
    ],

    "AUC":[
        cnn["AUC"][0],
        mobilenet["AUC"][0]
    ],

    "Training Time":[
        cnn["Training Time"][0],
        mobilenet["Training Time"][0]
    ],

    "Parameters":[
        cnn["Parameters"][0],
        mobilenet["Parameters"][0]
    ]

})

comparison
```

Accuracy Comparison

```python
plt.figure(figsize=(8,5))

plt.bar(
    comparison["Model"],
    comparison["Accuracy"]
)

plt.title("Accuracy Comparison")

plt.ylabel("Accuracy")

plt.grid(axis="y")

plt.show()
```

Precision Comparison

```python
plt.figure(figsize=(8,5))

plt.bar(
    comparison["Model"],
    comparison["Precision"]
)

plt.title("Precision Comparison")

plt.ylabel("Precision")

plt.grid(axis="y")

plt.show()
```

Recall Comparison

```python
plt.figure(figsize=(8,5))

plt.bar(
    comparison["Model"],
    comparison["Recall"]
)

plt.title("Recall Comparison")

plt.ylabel("Recall")

plt.grid(axis="y")

plt.show()
```

F1 Score Comparison

```python
plt.figure(figsize=(8,5))

plt.bar(
    comparison["Model"],
    comparison["F1 Score"]
)

plt.title("F1 Score Comparison")

plt.ylabel("F1 Score")

plt.grid(axis="y")

plt.show()
```

AUC Comparison

```python
plt.figure(figsize=(8,5))

plt.bar(
    comparison["Model"],
    comparison["AUC"]
)

plt.title("AUC Comparison")

plt.ylabel("AUC")

plt.grid(axis="y")

plt.show()
```

Temps d'entraînement

```python
plt.figure(figsize=(8,5))

plt.bar(
    comparison["Model"],
    comparison["Training Time"]
)

plt.title("Training Time")

plt.ylabel("Seconds")

plt.grid(axis="y")

plt.show()
```

Nombre de paramètres

```python
plt.figure(figsize=(8,5))

plt.bar(
    comparison["Model"],
    comparison["Parameters"]
)

plt.title("Number of Parameters")

plt.ylabel("Parameters")

plt.grid(axis="y")

plt.show()
```
